```python
start = 7200
import yfinance as yf
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

# 데이터를 가져오는 함수 정의
def fetch_history(ticker_str):
    ticker = yf.Ticker(ticker_str)
    hist = ticker.history(period="3d")
    if not hist.empty:
        hist["Ticker"] = ticker_str
        return hist
    return None

# 최대 10개의 스레드로 동시에 작업
with ThreadPoolExecutor() as executor:
    # 각 티커에 대해 fetch_history 함수를 실행하도록 작업을 제출
    future_to_ticker = {executor.submit(fetch_history, ticker): ticker for ticker in ticker_list[start:start+200]}
    
    # 작업이 완료되는 순서대로 결과 처리
    for future in as_completed(future_to_ticker):
        result = future.result()
        if result is not None:
            us_df = pd.DataFrame(result)
            us_df.reset_index(inplace=True)
            us_df = us_df[["Date", "Volume", "Ticker"]]
            us_history = pd.concat([us_history, us_df])
            print(f"{us_df['Ticker'].iloc[0]} done")
```
